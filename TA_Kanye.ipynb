```python
import pandas as pd
```

```python
df = pd.read_csv('kanye_west_lyrics.csv')
```

```python
df.head()
```

```python
songs = df.groupby('song').agg({'lyric':lambda x: " ".join(x), 'year':'mean'}).reset_index()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
```

```python
stop_words.extend(['back','said','come','things','get','oh','one','yeah','place','would','like','know','stay','go','let','cause','child','night','nothing','nanna'])
```

```python
vectorizer = TfidfVectorizer(stop_words = stop_words,min_df = 0.1)
```

```python
tfidf = vectorizer.fit_transform(songs['lyric'])
```

```python
from sklearn.decomposition import NMF
```

```python
nmf =  NMF(n_components=10)
```

```python
topic_values = nmf.fit_transform(tfidf)
```

```python
for topic_num,topic in enumerate(nmf.components_ ):
    message = " Topic #{}: ".format(topic_num +1)
    message += " ".join([vectorizer.get_feature_names()[i]for i in topic.argsort()[:-10:-1]])
```

```python
print(message)
```

```python
topic_labels = ['love','hoes','dark','gold-digger','pablo','Kim-kardashian','wife','hood','mula','youngmoney']
```

```python
df_topics = pd.DataFrame(topic_values,columns = topic_labels)
```

```python
df_topics
```

```python
songs = songs.join(df_topics)
```
